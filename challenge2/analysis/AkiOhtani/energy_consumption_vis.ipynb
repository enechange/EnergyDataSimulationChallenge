# 2. visualization of the data-set as values per day

created by the kmnmy(南谷 和毅)

```python
colors = ['#FFEE99','#FF9922','#FF4422']
```

```python
myDate = pd.date_range('2011-04-18', '2011-05-24')
energy_per_8hour = np.zeros([4, len(myDate)])
for i, one_date in enumerate(myDate[:]):
    one_date = one_date.strftime('%Y-%m-%d')
    energy_per_8hour[0,i] = dataFrame[one_date:one_date+' 08:00:00'].sum()
    energy_per_8hour[1,i] = dataFrame[one_date+' 08:00:00':one_date+' 16:00:00'].sum()
    energy_per_8hour[2,i] = dataFrame[one_date+' 16:00:00':one_date].sum()
    energy_per_8hour[3,i] = dataFrame[one_date].sum()
ind = np.arange(len(myDate))
width = 0.7
plt.figure(figsize=(16, 6))    
p1 = plt.bar(ind, energy_per_8hour[0], width, color=colors[0])
p2 = plt.bar(ind, energy_per_8hour[1], width, color=colors[1], bottom=energy_per_8hour[0])
p3 = plt.bar(ind, energy_per_8hour[2], width, color=colors[2], bottom=energy_per_8hour[0]+energy_per_8hour[1])
plt.xlabel('Day')
plt.ylabel('Total Energy Consumption in a day (W)')
plt.title('Energy Consumption per day from April 18 to May 24 in 2011*')
plt.xticks(ind+width/2., [dt.strftime(i, '%m/%d') for i in myDate], rotation=50)
plt.legend( (p3[0], p2[0], p1[0]), ('16:00-24:00', '8:00-16:00', '00:00-8:00'))
plt.autoscale(enable=True, axis='x',tight=True)
plt.show()

### Calculating Pearson correlation coefficients between energy consumption of three groups with it of all day
energy_per_8hour=np.delete(energy_per_8hour, [0,16,17,18,36], 1)
corr = np.corrcoef([energy_per_8hour[0], energy_per_8hour[1], energy_per_8hour[2], energy_per_8hour[3]])
print 'The Pearson correlation coefficients between energy consumption of three groups and energy consumption of all day'
print '    Time    '+ ' PCC'
print '00:00-08:00 '+ str(round(corr[3,0], 3))
print '08:00-16:00 '+ str(round(corr[3,1], 3))
print '16:00-24:00 '+ str(round(corr[3,2], 3))
```

# 3. visualization of the data-set as clusters : low, medium, and high energy consumption

created by the kmnmy(南谷 和毅)

```python
myDays = pd.Series('Mon Tue Wed Thu Fri Sat Sun'.split())
df_energy_per_day = dataFrame.resample('d', how='mean')
ward = AgglomerativeClustering(n_clusters=3, linkage='ward')
ward.fit([[x] for x in df_energy_per_day.Energy.fillna(0).tolist()])
df_energy_per_day['label'] = ward.labels_    

# For sorting label (;;)
depd = df_energy_per_day
label_sorted = sorted([depd.ix[depd.label == i].ix[0][0] for i in range(3)])
color_sorted = [0]*3
for i, label in enumerate(map(lambda x: depd.ix[depd.Energy == x].ix[0][1], label_sorted)):
    color_sorted[int(label)] = colors[i]

plt.figure(figsize=(16, 6))
for day in df_energy_per_day.groupby(lambda x: x.weekday):
    for row in range(day[1].Energy.count()):
        color = color_sorted[int(day[1].iloc[[row]].label)]
        plt.bar(0.1+ day[0] + float(row)/7, day[1].iloc[[row]].Energy, 0.12, color=color)

plt.xticks(np.arange(0.5,7.5,1), myDays)
for x in [1.03,2.01,2.95,3.95,4.95,5.95]:
    plt.axvline(x=x, linestyle='--', color='grey')
plt.xlabel('day of the week')
plt.ylabel('Energy Consumption in a day')
plt.title('Difference of Energy Consumption per day of the week*')

p = [0]*3
for i in range(3):
    p[i] = plt.Rectangle((0, 0), 1, 1, fc=colors[i])
plt.legend([p[2],p[1],p[0]], ["high","medium","low"], loc=2)
plt.show()
```

```python
energy_per_time=[]
for h in range(24):
    energy_per_time.append(dataFrame.at_time('%d:22:00'%(h)).Energy.mean())
    energy_per_time.append(dataFrame.at_time('%d:52:00'%(h)).Energy.mean())
N=24
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
width = np.ones(N)*2*np.pi/N
std=np.std(np.array(energy_per_time))
mean=np.array(energy_per_time).mean()

plt.figure(figsize=(16, 6))

radii_am = np.r_[energy_per_time[5::-1], energy_per_time[23:5:-1]]

ax_am = plt.subplot(1,2,1, polar=True)
plt.thetagrids(np.arange(0, 360, 360/12), labels=[3,2,1,0,11,10,9,8,7,6,5,4])
ax_am.set_title("Average Consumption per 30 mins from 00:00 to 12:00*", va='bottom')
bars_am = ax_am.bar(theta, radii_am, width=width, bottom=0)

radii_pm =  np.r_[energy_per_time[29:23:-1], energy_per_time[47:29:-1]]
ax_pm = plt.subplot(1,2,2, polar=True)
plt.thetagrids(np.arange(0, 360, 360/12), labels=[15,14,13,12,23,23,21,20,19,18,17,16])
ax_pm.set_title("Average Consumption per 30 mins (PM)*", va='bottom')
bars_pm = ax_pm.bar(theta, radii_pm, width=width, bottom=0)

std=np.std(np.array(energy_per_time))
mean=np.array(energy_per_time).mean()
for r, bar in zip(radii_am, bars_am)+ zip(radii_pm, bars_pm):
    if mean+std<r: bar.set_facecolor(colors[2])
    elif mean-std<r<mean + std: bar.set_facecolor(colors[1])
    else: bar.set_facecolor(colors[0])

plt.legend([p[2],p[1],p[0]], ["high( > mean+std)","medium","low( < mean-std)"], bbox_to_anchor=(0.85, 0.15), loc=2)
plt.show()

```

```python
days_and_time = np.zeros([7, 24])
for day in np.arange(0,7,1):
    for time in np.arange(0,24,1):
        days_and_time[day][time]=dataFrame.ix[(dataFrame.Energy.weekday == day) & (dataFrame.index.hour == time)].mean()
    
fig, ax = plt.subplots(figsize=(16,6))
cMap = ListedColormap([colors[0],colors[1],colors[2]])
heatmap = ax.pcolor(days_and_time, cmap=cMap,edgecolors='w')
cbar = plt.colorbar(heatmap)
cbar.ax.get_yaxis().set_ticks([])

for j, lab in enumerate(['low','medium','high']):
    cbar.ax.text(.5, (2 * j + 1) / 6.0, lab, ha='center', va='center', rotation=90)
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Clustering of Energy Consumption', rotation=90)

ax.set_xticks(np.arange(days_and_time.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(days_and_time.shape[0])+0.5, minor=False)
    
ax.invert_yaxis()
ax.xaxis.tick_top()
ax.axis('tight')
ax.set_xticklabels(np.arange(0,24,1), minor=False)
ax.set_yticklabels(myDays, minor=False)
plt.xlabel('Time')
plt.ylabel('Day of the week')
plt.text(0.5, 1.08, 'Relationship of Energy Consumption between time and day of the week*',
        horizontalalignment='center',
        fontsize=18,
        transform = ax.transAxes)
plt.show()

```

<h2>Sunburst Trees</h2>
<p>This is an example of using <a href="http://github.com/mbostock/d3/wiki/Partition-Layout">d3.layout.partition</a> to generate a zoomable sunburst tree derived from hierarchical data.  A <a href="http://www.cc.gatech.edu/gvu/ii/sunburst/">sunburst tree</a> is a radial space-filling visualisation, analagous to an icicle tree.
<h2>Colours</h2>
<p>The original colours appear to have been selected by the designer, as opposed to matching any standard colour name palettes.  So I painstakingly copied the colours from the outermost (leaf) nodes of the tree to match the original.
<p>The inner node colours are calculated automatically by taking the average of the two child nodes in HSL colour space.
<h2>Labels</h2>
<p>The label colour of black or light grey is automatically chosen by calculating the <a href="http://www.w3.org/WAI/ER/WD-AERT/#color-contrast">colour brightness</a> of each node.  If a label has more than one word, it is split into two lines and placed equidistant to the centre of the segment.
